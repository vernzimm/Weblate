```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
import os

a = os.getcwd()
a = a + "\\teeth.txt"
print(a)
```

```python
b = open(a)

x = []
for i in b :
    i = i.split()
    x = x + [[float(i[0]),float(i[1]),float(i[2])]]
b.close

z = np.array(x)

phi = np.zeros((0,3))
for i in z :
    a = np.sqrt(i[0]**2 + i[1]**2)
    b = np.arctan2(i[1],i[0]) * 180 / np.pi
    phi = np.append(phi,[[a, b, i[2]]], axis = 0)
```

```python
predges = np.linspace(17,20,30)
paedges = np.linspace(-15,15,100)
zedges = np.linspace(-10,0,2)

hist, edges = np.histogramdd(phi, bins = (predges,paedges,zedges))
pridx, paidx, zidx = np.where(hist>0)
plt.scatter(predges[pridx], paedges[paidx], c = zedges[zidx], alpha=0.7)
plt.show()

#a = plt.scatter(z[0],z[1])
#plt.show()
```
